```python
#importing NumPy library for arrays and computation
import numpy as np 
#importing matplotlib library for plotting
import matplotlib.pyplot as plt

#setting style of plots
plt.style.use('seaborn') 
```

```python
#creating an array of input features
x_train = np.array([1.0, 2.0])
#creating an array of output targets (training set)
y_train = np.array([300.0, 500.0])

#f-strings allow embedding variables in strings directly
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
#shape property gives a tuple of dimensions of the array
print(f"x_train.shape : {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
i = 1
x_i = x_train[i]
y_i = y_train[i]
print(f"x^({i}),y^({i}) = ({x_i},{y_i})")
```

```python
#scatter plots the points, marker = x -> cross shape, c = r -> red colour
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 200
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
#function to computer linear regression output (predicted line to fit the data model)
def compute_model_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train,w,b)

#plotting our predicted regression line
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
#Now we can predict cost of any house (E.g :- Predict cost of 1200 sq. ft house)
x_i = 1.2
f_wb = w * x_i + b

print(f"Cost of 1200 sq.ft house = {f_wb*1000} $")
```
